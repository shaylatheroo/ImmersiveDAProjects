```python
import json
import pandas as pd

filepath = 'MoCoViolations_2018.json'
```

```python
data = []


# Each line inside Json file doesn;t have a separator like comma
with open(filepath, 'r') as file:
    vio_data = json.load(file)
```

```python
len(vio_data)
```

```python
vio_data.keys()
```

```python
DateofStop=[]
TimeofStop=[]
Agency=[]
Subagency=[]
Description=[]
Location=[]
Latitude=[]
Longitude=[]
VehicleType=[]
VehicleYear=[]
VehicleMake=[]
VehicleModel=[]
VehicleColor=[]
ViolationType=[]
Race=[]
Gender=[]
DriversState=[]
DriversLicenseState=[]
ArrestType=[]
```

```python
currentRecord = 0
for x in vio_data['data']:
    DateofStop.append(vio_data['data'][currentRecord][9])
    TimeofStop.append(vio_data['data'][currentRecord][10])
    Agency.append(vio_data['data'][currentRecord][11])
    Subagency.append(vio_data['data'][currentRecord][12])
    Description.append(vio_data['data'][currentRecord][13])
    Location.append(vio_data['data'][currentRecord][14])
    Latitude.append(vio_data['data'][currentRecord][15])
    Longitude.append(vio_data['data'][currentRecord][16])
    VehicleType.append(vio_data['data'][currentRecord][28])
    VehicleYear.append(vio_data['data'][currentRecord][29])
    VehicleMake.append(vio_data['data'][currentRecord][30])
    VehicleModel.append(vio_data['data'][currentRecord][31])
    VehicleColor.append(vio_data['data'][currentRecord][32])
    ViolationType.append(vio_data['data'][currentRecord][33])
    Race.append(vio_data['data'][currentRecord][37])
    Gender.append(vio_data['data'][currentRecord][38])
    DriversState.append(vio_data['data'][currentRecord][40])
    DriversLicenseState.append(vio_data['data'][currentRecord][41])
    ArrestType.append(vio_data['data'][currentRecord][42])
    currentRecord += 1
```

```python
len(DateofStop)
```

```python
df=pd.DataFrame(DateofStop, columns=['DateofStop'])

#add other lists as new columns in my new dataframe
df['TimeofStop'] = TimeofStop
df['Agency'] = Agency
df['Subagency'] = Subagency
df['Description'] = Description
df['Location'] = Location
df['Latitude'] = Latitude
df['Longitude'] = Longitude
df['VehicleType'] = VehicleType
df['VehicleYear'] = VehicleYear
df['VehicleMake'] = VehicleMake
df['VehicleModel'] = VehicleModel
df['VehicleColor'] = VehicleColor
df['ViolationType'] = ViolationType
df['Race'] = Race
df['Gender'] = Gender
df['DriversState'] = DriversState
df['DriversLicenseState'] = DriversLicenseState
df['ArrestType'] = ArrestType

#show first 5 rows of created dataframe
df.head()
```
